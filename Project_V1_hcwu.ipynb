```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Input, Sequential
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split

# -----------------------------
# 1. Load and Inspect Data
# -----------------------------
df = pd.read_csv('signals_exported.csv')
time = df.iloc[:, 0]
dt = time.diff().median()  # sampling interval
window_size = int(1.0 / dt)   # number of samples in 1 second
horizon = int(0.001 / dt)     # number of samples in 0.001 second

print(f"Sampling interval (dt): {dt} s")
print(f"Window size (1 s): {window_size} samples")
print(f"Horizon (0.001 s): {horizon} samples ahead\n")

# -----------------------------
# 2. Define Features & Targets
# -----------------------------
feature_cols = [
    "Stator_voltage__vq__pu_", "Output_active_power___Peo__pu_", "vt_mag_2",
    "Stator_voltage__vd__pu_", "vt_mag_1", "Field_current__ifd__pu_", "Rotor_speed__wm__pu_"
]
target_cols = ["throttle", "Efd"]

features = df[feature_cols].values
targets = df[target_cols].values

# -----------------------------
# 3. Scale Features & Targets
# -----------------------------
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()

features_scaled = feature_scaler.fit_transform(features)
targets_scaled = target_scaler.fit_transform(targets)

# -----------------------------
# 4. Create Sequences
# -----------------------------
X, Y = [], []
for i in range(window_size, len(df) - horizon):
    X.append(features_scaled[i - window_size : i])
    Y.append(targets_scaled[i + horizon])

X = np.array(X)
Y = np.array(Y)

print("X shape:", X.shape)
print("Y shape:", Y.shape, "\n")

# -----------------------------
# 5. Train/Test Split
# -----------------------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, shuffle=False
)

# -----------------------------
# 6. Build LSTM Model
# -----------------------------
model = Sequential([
    Input(shape=(window_size, len(feature_cols))),
    LSTM(64),
    Dense(32, activation='relu'),
    Dense(len(target_cols))
])
model.compile(optimizer='adam', loss='mse')
model.summary()

# -----------------------------
# 7. Train
# -----------------------------
history = model.fit(
    X_train, Y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.1
)

import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error Loss')
plt.legend()
plt.grid(True)
plt.show()


# -----------------------------
# 8. Evaluate & Predict
# -----------------------------
test_loss = model.evaluate(X_test, Y_test)
print("\nTest Loss (MSE):", test_loss)

Y_pred_scaled = model.predict(X_test)
Y_pred = target_scaler.inverse_transform(Y_pred_scaled)
Y_true = target_scaler.inverse_transform(Y_test)

# Optional: quick look at first few predictions vs. true values
print("\nFirst 5 Predictions vs True:")
for i in range(5):
    print(f"Predicted: {Y_pred[i]},  True: {Y_true[i]}")
```
